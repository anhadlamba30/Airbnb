```python
import pandas as pd
import numpy as np 
```

```python
#importing data
airbnb_data_path = "../Data/Airbnb_Data.csv"
df = pd.read_csv(airbnb_data_path)
```

```python
df.head()
```

```python
df.describe()
```

```python
cols = df.columns
print(cols)

# droping columns that are not useful
df.drop(columns = ['id','description','host_has_profile_pic',
                   'latitude','longitude','neighbourhood',
                   'first_review','last_review','accommodates',
                   'host_response_rate','thumbnail_url','name',
                   'amenities','host_since','zipcode'
                  ],inplace = True)
```

```python
import pandas as pd


df = df.dropna(subset=['host_identity_verified'])

# Update 'instant_bookable' column
df['instant_bookable'] = df['instant_bookable'].map({'t': 1, 'f': 0})

# Update 'host_identity_verified' column
df['host_identity_verified'] = df['host_identity_verified'].map({'t': 1, 'f': 0})
```

```python
from sklearn.preprocessing import OneHotEncoder

# Instantiate the encoder
encoder = OneHotEncoder()

# Fit and transform the categorical columns
array_hot_encoded = encoder.fit_transform(df[['property_type', 'room_type', 'bed_type', 'cancellation_policy','city']])

# Convert the array to a DataFrame
data_hot_encoded = pd.DataFrame(array_hot_encoded.toarray(), columns=encoder.get_feature_names_out(['property_type', 'room_type', 'bed_type', 'cancellation_policy','city']), index=df.index)

# Concatenate the encoded columns with the original data
df_encoded = pd.concat([data_hot_encoded, df.drop(columns=['property_type', 'room_type', 'bed_type', 'cancellation_policy','city'])], axis=1)

# Now 'data_out' contains the one-hot encoded columns
df_encoded.head()
```

```python
df_encoded = df_encoded.dropna(subset=['bathrooms','review_scores_rating','bedrooms','beds'])
```
